```python
import pandas as pd
import numpy as np
import scipy as sp
from scipy.optimize import linprog
from copy import copy
```

```python
k = 4
l = 7
n = 10 + k//4
m = 30 - k//4
b = [70 + l + k + i for i in range(m)]
c = [50 + l + k - i for i in range(n)]
A = [[1 + ((j + k) * i + j**2 + i**3 + 3*(i + l)) % (30 + k//5) for i in range(n)] for j in range(m)]
```

```python
def display(A, b):
    for i in range(m):
        for j in range(n):
            print(f"{A[i][j]}*x_{j}", end='\t')
        print(f"<= {b[i]}")
    print("\n")
```

```python
display(A,b)
```

```python
c
```

```python
res = linprog([-i for i in c], A_ub=A, b_ub=b, method='simplex', options={"disp": True})
```

```python
res
```

```python
sum(list(res.x * np.array(c)))
```

```python
def proverka(x):
    res = []
    for i in range(m):
        res.append(sum([A[i][j] * x[j] for j in range(n)]) <= b[i])
        print(sum([A[i][j] * x[j] for j in range(n)]),'<=', b[i])
    return all(res)
```

```python
proverka(res.x)
```

```python
h = 0
best = -1
for k in range(m):
    b_piv = copy(b)
    b_piv[k] += 1
    res = linprog([-i for i in c], A_ub=A, b_ub=b_piv, method='simplex', options={"disp": False})
    if sum(list(res.x * np.array(c))) > h:
        best = k
        h = sum(list(res.x * np.array(c)))
h, best
```

```python
b_piv = copy(b)
b_piv[best+1] += 1
res = linprog([-i for i in c], A_ub=A, b_ub=b_piv, method='simplex', options={"disp": True})
```

```python
res.x
```

```python
sum(list(res.x * np.array(c)))
```

```python
proverka(res.x)
```
