```python
import gurobipy as gp
from gurobipy import GRB
import numpy as np
```

Считаем переменные по моему варианту и заполняем матрицы

```python
k=4
l=7
x = k//4
n=10+x
m=30-x
```

```python
C
```

```python
for i in range(m):
    B[i] = 70+l+k+i
print(B)
```

```python
for i in range(n):
    C[i] = 50+l+k-i
print(C)
```

```python
x = k/5
x= round(x)
for i in range(m):  
  for j in range(n):
      A[i] = 1+((j+k)*i+j*j+i*i*i+3*(i+l)) % (30+x)
print(A)
```

Создаем модель для оптимизации целевой функции и запускаем её

```python
# Declare and initialize model
m1 = gp.Model('RAP')
x = m1.addMVar(shape=13, vtype=GRB.BINARY, name="x")
m1.setObjective(C @ x, GRB.MAXIMIZE)
m1.addConstr(A @ x <= B, name="c")
m1.optimize()
```

```python
 print(x.X)
```

```python
print('Obj: %g' % m1.ObjVal)
```

Создаем модель для оптимизации двойственной задачи и запускаем её

```python
m2 = gp.Model('RAP2')
y = m2.addMVar(shape=13, vtype=GRB.BINARY, name="y")
m2.setObjective(C @ x, GRB.MINIMIZE)
m2.addConstr(A @ x <= B, name="c")
m2.optimize()
```

```python
 print(y.X)
```

```python
print('Obj: %g' % m2.ObjVal)
```
